Our next aim is to create a dataframe that contains 3 columns that are movie id ,  title of the movie and the tag of the movie. The tag column contains all the information about the movie that is about the overview of the movie, from which genre does the movie belong to , the main charaters in the movie and finally the crew memebers like the director name

```python
# [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}] -------->1
# ['Action' , 'Adventure' , 'Fantasy' , 'Scienece Fiction'] ------>2

# We have the genre in  the format like number 1 which is not favourble format in order to merge in the tag column. So in order carry out the process we need to genre column from format 1 to format 2
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# In order to covert all the rows of genres from format1 to format 2 we need to call a function which will make our work easier. But there is a problem , the genre column has the string of list which is not acceptable. So the ast finction used to convert the string of list into a list
# Passing a funtion to only extract the name os the different types pf genres mentioned in the dataframe

def convert(obj):
  L =[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies.genres = movies.genres.apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0].keywords
```

```python
movies.keywords = movies.keywords.apply(convert)
```

```python
movies.head()
```

```python
#Function to extract only the main characters of the movie that the first 3 main characters that are mentioned in the dataframe

def convert_cast(obj):
  L =[]
  c=0
  for i in ast.literal_eval(obj):
    if(c<3):
      L.append(i['name'])
      c+=1
    else:
      break
  return L
```

```python
movies.cast = movies.cast.apply(convert_cast)
```

```python
movies.head()
```

```python
movies.crew[0]
```

```python
# Passing a function to only extract the name of the directors of the moovies mentioned in the dataframe

def convert_crew(obj):
  l = []
  for i in ast.literal_eval(obj):
    if(i['job']== 'Director'):
      l.append(i['name'])
    else:
      continue
  return l
```

```python
movies.crew = movies.crew.apply(convert_crew)
```

```python
movies.head()
```

```python
movies.overview = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# 'Sam Worthington' ------> 'SamWorthington'
# We are carrying out this process because when our system will making tags then it will make two different tags for the name and the surname will may later confuse the system since two diffrent person might habe same name. For this reason system migth recommend a wrong output or movie on which the user in not so interested

movies.genres = movies.genres.apply(lambda x:[i.replace(" ","") for i in x])
movies.keywords = movies.keywords.apply(lambda x:[i.replace(" ","") for i in x])
movies.crew = movies.crew.apply(lambda x:[i.replace(" ","") for i in x])
movies.cast = movies.cast.apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies.overview +movies.genres + movies.keywords + movies.cast + movies.crew
```

```python
movies.head()
```

```python
movies = movies.drop(['overview','keywords','cast','crew','genres'],axis=1)
```

```python
movies.head()
```

```python
movies.tags = movies.tags.apply(lambda x: " ".join(x))
movies.head()
```

```python
movies.tags = movies.tags.apply(lambda x : x.lower())
```

```python
movies.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
movies.tags = movies.tags.apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# cv = CountVectorizer()
res = CountVectorizer(max_features = 5000 , stop_words = 'english')
```

```python
vectors = res.fit_transform(movies.tags).toarray()
vectors
```

```python
res.get_feature_names_out()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
similarity[0] , similarity[0].shape
```

```python
def recommend(movie):
  movie_index  = movies[movies.title == movie].index[0]
  distances = similarity[movie_index]
  movie_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])

  for i in movie_list [1:6]:
    print(movies.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
